### Preprocesamiento de Datos e Ingenieria de Caracteristicas

* Concatenamos las columnas que se repiten en ambos DataFrames. Esto con el fin de ir creando el conjunto de datos enriquecido.

```python
df_combinate = pd.concat([df_leads[["Id", "Status", "Converted", "Use Case", "Created Date"]], df_offers[["Id", "Status", "Use Case", "Created Date"]]]).reset_index(drop=True)
```

```python
df_combinate
```

* Modificamos Id's repetidos y nulos (NaN) con valores nuevos con las mismas caracteristicas las cuales son:
    * valores aleatorios que contengan letras y n√∫meros.
    * Maxima cantidad 8 digitos.

```python
# Asignar un valor temporal a NaN para evitar el error
df_combinate['Id'].fillna('temp', inplace=True)

# Obtener los valores que se repiten
repetidos = df_combinate['Id'].value_counts()[df_combinate['Id']].gt(1).values

# Generar nuevos valores aleatorios para los valores repetidos y NaN
nuevos_valores = np.random.choice(list('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'), (len(df_combinate), 8))
df_combinate['Id'] = np.where(repetidos | (df_combinate['Id'] == 'temp') | df_combinate['Id'].isna(), [''.join(valor) for valor in nuevos_valores], df_combinate['Id'])

# Revertir los valores temporales a NaN
df_combinate['Id'].replace('temp', np.nan, inplace=True)
```

* En la columna **Converted** se va a llamar **Probability of conversion**.
  
  El orden con simbolos que hicimos anteriormente fue con el fin de relacionarlo con la nueva columna llamada **Probability of conversion**. Su relaci√≥n ser√≠a la siguiente:

  * Sabemos que la DataFrame **Leads** contiene los datos de todos los clientes potenciales. Por tanto su probabilidad de recibir el servicio (Probability of conversion) ser√° entre *0.0* y *0.5*

  * Sabemos que la DataFrame **Offers** contiene los datos de todos los clientes que asistieron al menos a la reuni√≥n de demostraci√≥n. Por tanto su probabilidad de recibir el servicio (Probability of conversion) ser√° entre *0.5* y *0.9*


Entonces la columna **Probability of conversion** tendra valores de *0.0* a *0.9* y va estar relacionado con la columna **Status Category** de la siguiente forma:

    Closed lost -> 0.0

    Discarded -> 0.1

    Recicled -> 0.2

    Process -> 0.3

    Waiting -> 0.4

    Engaged -> 0.45

    Qualified -> 0.5

    Demo 1 -> 0.6

    Demo 2 -> 0.7

    Checkbox -> 0.8

    Negotiation -> 0.85

```python
df_combinate = df_combinate.rename(columns={"Converted": "Probability of conversion"})
df_combinate["Probability of conversion"] = df_combinate["Status Category"].replace({
    "Closed Lost": 0.0,
    "Discarded": 0.1,
    "Recicled": 0.2,
    "Process": 0.3,
    "Waiting": 0.4,
    "Engaged": 0.45,
    "Qualified": 0.5,
    "Demo 1": 0.6,
    "Demo 2": 0.7,
    "Checkbox": 0.8,
    "Negotiation": 0.85,
    "Closed Won": 0.9
})
```
